```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# lista de palavras
palavras = ["exemplo", "palavra", "nuvem", "gerada", "python", "biblioteca", "wordcloud"]

# gerar a nuvem de palavras
nuvem_palavras = WordCloud().generate(' '.join(palavras))

# definir cores
cmap = plt.cm.gray
cores = [cmap(i) for i in range(len(nuvem_palavras.words_))]

# funo para exibir a nuvem de palavras
def exibir_nuvem(palavras, **kwargs):
    fig, ax = plt.subplots()
    nuvem_palavras = WordCloud(**kwargs).generate(' '.join(palavras))
    ax.imshow(nuvem_palavras, interpolation='bilinear')
    ax.axis("off")
    return fig, ax

# funo para destacar a palavra sob o mouse
def realcar_palavra(evento):
    if not evento.inaxes:
        return
    try:
        idx = nuvem_palavras.words_.index(evento.artist.get_text())
    except ValueError:
        return
    cores[idx] = cmap(0.6) # destaque em amarelo
    for i, palavra in enumerate(nuvem_palavras.words_):
        evento.artist.axes.text(nuvem_palavras.coordinates_[i][0], nuvem_palavras.coordinates_[i][1], palavra, 
                                color=cores[i], fontdict=nuvem_palavras.font_path,
                                rotation=nuvem_palavras.orientation_[i], fontsize=nuvem_palavras.sizes[i])
    evento.canvas.draw_idle()

# funo para restaurar as cores originais
def restaurar_cores(evento):
    cores[:] = [cmap(i) for i in range(len(nuvem_palavras.words_))]
    for i, palavra in enumerate(nuvem_palavras.words_):
        evento.artist.axes.text(nuvem_palavras.coordinates_[i][0], nuvem_palavras.coordinates_[i][1], palavra, 
                                color=cores[i], fontdict=nuvem_palavras.font_path,
                                rotation=nuvem_palavras.orientation_[i], fontsize=nuvem_palavras.sizes[i])
    evento.canvas.draw_idle()

# exibir a nuvem de palavras interativa
fig, ax = exibir_nuvem(palavras)
fig.canvas.mpl_connect('motion_notify_event', realcar_palavra)
fig.canvas.mpl_connect('axes_leave_event', restaurar_cores)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import numpy as np

# lista de palavras
palavras = ["exemplo", "palavra", "nuvem", "gerada", "python", "biblioteca", "wordcloud"]

# gerar a nuvem de palavras
nuvem_palavras = WordCloud().generate(' '.join(palavras))

# definir cores
cmap = plt.cm.gray
cores = [cmap(i) for i in range(len(nuvem_palavras.words_))]

# funo para exibir a nuvem de palavras
def exibir_nuvem(palavras, **kwargs):
    fig, ax = plt.subplots()
    nuvem_palavras = WordCloud(**kwargs).generate(' '.join(palavras))
    ax.imshow(nuvem_palavras, interpolation='bilinear')
    ax.axis("off")
    return fig, ax

# funo para destacar a palavra sob o mouse
def realcar_palavra(evento):
    if not evento.inaxes:
        return
    try:
        idx = nuvem_palavras.words_.index(evento.artist.get_text())
    except ValueError:
        return
    cores[idx] = cmap(0.6) # destaque em amarelo
    for i, palavra in enumerate(nuvem_palavras.words_):
        evento.artist.axes.text(nuvem_palavras.coordinates_[i][0], nuvem_palavras.coordinates_[i][1], palavra, 
                                color=cores[i], fontdict=nuvem_palavras.font_path,
                                rotation=nuvem_palavras.orientation_[i], fontsize=nuvem_palavras.sizes[i])
    evento.canvas.draw_idle()

# funo para restaurar as cores originais
def restaurar_cores(evento):
    cores[:] = [cmap(i) for i in range(len(nuvem_palavras.words_))]
    for i, palavra in enumerate(nuvem_palavras.words_):
        evento.artist.axes.text(nuvem_palavras.coordinates_[i][0], nuvem_palavras.coordinates_[i][1], palavra, 
                                color=cores[i], fontdict=nuvem_palavras.font_path,
                                rotation=nuvem_palavras.orientation_[i], fontsize=nuvem_palavras.sizes[i])
    evento.canvas.draw_idle()

# funo para embaralhar as palavras
def embaralhar_palavras(b):
    global palavras
    global nuvem_palavras
    global cores
    # embaralhar palavras
    np.random.shuffle(palavras)
    # regerar a nuvem de palavras
    nuvem_palavras = WordCloud().generate(' '.join(palavras))
    # regerar as cores
    cores = [cmap(i) for i in range(len(nuvem_palavras.words_))]
    # redesenhar a nuvem de palavras
    ax.clear()
    ax.imshow(nuvem_palavras, interpolation='bilinear')
    ax.axis("off")
    fig.canvas.draw_idle()

# criar boto para embaralhar as palavras
botao_embaralhar = widgets.Button(description="Embaralhar Palavras")
botao_embaralhar.on_click(embaralhar_palavras)
display(botao_embaralhar)

# exibir a nuvem de palavras interativa
fig, ax = exibir_nuvem(palavras)
```

```python
import networkx as nx
import plotly.graph_objects as go

G = nx.Graph()
G.add_nodes_from(['a', 'b', 'c', 'd', 'e'])
G.add_edges_from([('a', 'b'), ('a', 'c'), ('a', 'd'), ('b', 'd'), ('c', 'd'), ('d', 'e')])
```

```python
pos = nx.spring_layout(G)
for node in G.nodes:
    G.nodes[node]['pos'] = list(pos[node])
```

```python
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = G.nodes[edge[0]]['pos']
    x1, y1 = G.nodes[edge[1]]['pos']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

node_x = []
node_y = []
node_text = []
for node in G.nodes():
    x, y = G.nodes[node]['pos']
    node_x.append(x)
    node_y.append(y)
    node_text.append(node)
```

```python
fig = go.Figure(go.Scatter(x=edge_x,
                           y=edge_y,
                           mode='lines',
                           line=dict(color='rgb(125,125,125)', width=1),
                           hoverinfo='none'))

fig.add_trace(go.Scatter(x=node_x,
                         y=node_y,
                         mode='markers+text',
                         text=node_text,
                         hoverinfo='text',
                         marker=dict(symbol='circle-dot',
                                     size=30,
                                     color='#6175c1',
                                     line=dict(color='rgb(50,50,50)', width=1))))

fig.update_layout(title='Nuvem de Grafo',
                  titlefont_size=20,
                  showlegend=False,
                  hovermode='closest',
                  margin=dict(b=20,l=5,r=5,t=40),
                  xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                  yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))

fig.show()
```

```python
import random
import plotly.graph_objs as go
from wordcloud import WordCloud, STOPWORDS

palavras = {'python': 200, 'data': 120, 'analyst': 100, 'machine': 80, 'learning': 60, 'big': 40, 'data': 30, 'analytics': 20, 'artificial': 10, 'intelligence': 10}
```

```python
def gerar_nuvem(palavras, hover_text):
    wc = WordCloud(width=800, height=400, background_color='black', max_words=200, stopwords=STOPWORDS)
    wc.generate_from_frequencies(palavras)

    colors = [random.randint(0, 255) for _ in range(len(palavras))]
    hoverinfo = 'text'

    trace = go.Scatter(x=[random.uniform(-1, 1) for _ in range(len(palavras))],
                       y=[random.uniform(-1, 1) for _ in range(len(palavras))],
                       mode='text',
                       text=list(palavras.keys()),
                       hovertext=hover_text,
                       hoverinfo=hoverinfo,
                       marker=dict(color=colors, size=15, line=dict(width=1, color='rgb(0, 0, 0)')),
                       textfont=dict(size=list(palavras.values()), color='white'),
                       textposition='middle center')

    return go.Figure(data=[trace],
                     layout=go.Layout(title=dict(text='Nuvem de Palavras'),
                                      xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                                      yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                                      hovermode='closest'))
```

```python
def realcar_palavra(trace, points, state):
    if points.point_inds:
        selecionados = points.point_inds
        cores = ['rgb(125, 125, 125)'] * len(palavras)
        for i in selecionados:
            cores[i] = 'rgb(255, 0, 0)'
        trace.marker.color = cores
```

```python
def restaurar_cores(trace, points, state):
    trace.marker.color = [random.randint(0, 255) for _ in range(len(palavras))]
```

```python
hover_text = [f'Frequncia: {frequencia}' for frequencia in palavras.values()]

fig = gerar_nuvem(palavras, hover_text)
fig.data[0].on_hover(realcar_palavra)
fig.data[0].on_unhover(restaurar_cores)
fig.show()
```
